```python
import tensorflow as tf
import numpy as np
import os
import csv
import pandas as pd
from pandas import DataFrame as df

train_path="./train.csv"
test_path="./test.csv"

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
train = train.drop(['Name','Cabin','Ticket','PassengerId'], axis=1)

train.loc[train['Sex']=='male', 'Sex']=1
train.loc[train['Sex']=='female', 'Sex']=2

train.loc[train['Embarked']=='C', 'Embarked']=1
train.loc[train['Embarked']=='Q', 'Embarked']=2
train.loc[train['Embarked']=='S', 'Embarked']=3

train=train.fillna(train.mean())
```

```python
test = test.drop(['Name','Cabin','Ticket','PassengerId'], axis=1)

test.loc[test['Sex']=='male', 'Sex']=1
test.loc[test['Sex']=='female', 'Sex']=2

test.loc[test['Embarked']=='C', 'Embarked']=1
test.loc[test['Embarked']=='Q', 'Embarked']=2
test.loc[test['Embarked']=='S', 'Embarked']=3

test=test.fillna(test.mean())
```

```python
test['Age'] /= test['Age'].mean()
test['Fare'] /= test['Fare'].mean()
```

```python
train['Age'] /= train['Age'].mean()
train['Fare'] /= train['Fare'].mean()
```

```python
train
```

```python
train_size=700
train_set = train.to_numpy()
x=[]
y=[]
for row in train_set:
  x.append(row[1:])
  y.append(row[0])


train_x = np.array(x[:train_size])
train_y = np.array(y[:train_size])
valid_x = np.array(x[train_size:])
valid_y = np.array(y[train_size:])


```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(train_x,train_y, epochs=500, validation_data = (valid_x, valid_y))
```

```python
y_predict = model.predict(test)
```

```python
with open("result.csv",'w') as file:
  writer = csv.writer(file)
  writer.writerow(['PassengerId','Survived'])
  for i in range(892, 892+len(y_predict)):
    label = 1 if y_predict[i-892] >0.5 else 0
    writer.writerow([i,label])
  
```
